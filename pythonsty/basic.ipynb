你就需要错误分支

> It is easier to ask for forgiveness than permission  
> —— Python 之父 Guido van Rossum

```python
try:
      pass
except:
      pass # 出错做什么
else:
      # hahhahahahahhahhahhahhhahhahhhahhahhahhahhahhahhha
finally:
      pass # 无论什么都做
```


````python
try:
    a = 1 + 1
except ZeroDivisionError as e:
    print(e)
else:
    print(a)
finally:
    print(12)
try:
    a = 1 / 0
except ZeroDivisionError as e:
    print(e)
else:
    print(a)
finally:
    print(13)
try:
    raise KeyboardInterrupt("我爱你™")
except (KeyboardInterrupt, ZeroDivisionError) as e:
    print("Ctrl+C has press", e)
finally:
    print(14)
try:
    a = 1 / 0
except KeyboardInterrupt as e:
    print(e)
finally:
    print(42)
````
