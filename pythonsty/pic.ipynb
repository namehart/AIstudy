## 绘图


```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

fig = plt.figure(figsize=(8, 6))

# Generate sample data for a sine wave
x = np.linspace(0, 3)
y = np.sin(x)
# Plot the data
plt.plot(x, y)
# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Hello World!')
```
